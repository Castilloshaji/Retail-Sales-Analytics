```python
## Handling Missing Values

### Observation
#The `Postal Code` column contains a large number of missing values (approximately 80%).

### Decision
#The column is retained because the missing values are inherent to the Global Superstore dataset, where many international orders do not have postal codes available.

#No imputation was performed to avoid introducing inaccurate geographic information. Missing values are preserved as `NULL`, allowing downstream systems such as PostgreSQL and Power BI to handle them appropriately.
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)
```

```python
df.columns
```

```python
df.dtypes
```

```python
df[["order_date", "ship_date"]].head()
```

```python
df["order_date"] = pd.to_datetime(
    df["order_date"],
    origin="1899-12-30",
    unit="D"
)

df["ship_date"] = pd.to_datetime(
    df["ship_date"],
    origin="1899-12-30",
    unit="D"
)
```

```python
df[["order_date", "ship_date"]].head()
```

```python
print(f"Duplicate Rows: {df.duplicated().sum()}")
```

```python
df.isnull().sum()
```

```python
print("Negative Sales:", (df["sales"] < 0).sum())
print("Negative Quantity:", (df["quantity"] < 0).sum())
print("Negative Shipping Cost:", (df["shipping_cost"] < 0).sum())
```

```python
print("Negative Profit:", (df["profit"] < 0).sum())
```

```python
print("Minimum Discount:", df["discount"].min())
print("Maximum Discount:", df["discount"].max())
```

```python
invalid_discount = df[(df["discount"] < 0) | (df["discount"] > 1)]

print(f"Invalid Discount Records: {len(invalid_discount)}")
```

```python
invalid_dates = df[df["ship_date"] < df["order_date"]]

print(f"Invalid Date Records: {len(invalid_dates)}")
```

```python
categorical_columns = [
    "ship_mode",
    "segment",
    "market",
    "region",
    "category",
    "sub_category",
    "order_priority"
]

for col in categorical_columns:
    print(f"\n{col.upper()}")
    print(df[col].unique())
```

```python
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = df[col].str.strip()
```
